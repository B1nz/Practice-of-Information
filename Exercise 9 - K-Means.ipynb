```python
# k-means clustering algorithm

#
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np

# generate random data consisting of 2 groups
X= -2 * np.random.rand(100,2)
X1 = 1 + 2 * np.random.rand(50,2)
X[50:100, :] = X1
plt.scatter(X[0:49, 0],X[0:49 , 1], s = 50, c = 'b')
plt.scatter(X[50:100, 0], X[50:100, 1], s = 50, c = 'y')
plt.show()
```

```python
Kmean = KMeans(n_clusters=2)
Kmean.fit(X)
# finding the center of the clusters
centers = Kmean.cluster_centers_
'''
array([[-0.94665068, -0.97138368],
[ 2.01559419, 2.02597093]])
'''
plt.scatter(X[0:49, 0],X[0:49 , 1], s = 50, c = 'b')
plt.scatter(X[50:100, 0], X[50:100, 1], s = 50, c = 'y')

plt.scatter(centers[0,0], centers[0,1], s=200, c='g', marker='s')
plt.scatter(centers[1,0], centers[1,1], s=200, c='r', marker='s')
plt.show()

```

```python
# Testing the algorithm:
# getting the labels property of the K-means clustering example dataset; that is, how the data points are categorized into the two clusters.

Kmean.labels_
# predicting the cluster of a data point
sample_test=np.array([4.0,4.0])

# the test data point belongs to the 0 (green centroid) cluster
second_test=sample_test.reshape(1, -1) # reshape sample_test to a one row (column is set as unspecified value)
Kmean.predict(second_test)
```
