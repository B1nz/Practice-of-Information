```python
# k-means clustering algorithm

#
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np
```

```python
# generate random data consisting of 2 groups
X= -2 * np.random.rand(150,2)
X1 = 1 + 2 * np.random.rand(30,2)
X2 = 4 + 2 * np.random.rand(30,2)
X3 = 7 + 2 * np.random.rand(30,2)
X4 = 10 + 2 * np.random.rand(30,2)
X[30:60, :] = X1
X[60:90, :] = X2
X[90:120, :] = X3
X[120:150, :] = X4
plt.scatter(X[0:30, 0],X[0:30 , 1], s = 50, c = 'r')
plt.scatter(X[30:60, 0], X[30:60, 1], s = 50, c = 'y')
plt.scatter(X[60:90, 0], X[60:90, 1], s = 50, c = 'g')
plt.scatter(X[90:120, 0], X[90:120, 1], s = 50, c = 'c')
plt.scatter(X[120:150, 0], X[120:150, 1], s = 50, c = 'b')
plt.show()
```

```python
Kmean = KMeans(n_clusters=5)
Kmean.fit(X)
# finding the center of the clusters
centers = Kmean.cluster_centers_
'''
array([[-0.94665068, -0.97138368],
[ 2.01559419, 2.02597093]])
'''
plt.scatter(X[0:30, 0],X[0:30 , 1], s = 50, c = 'r')
plt.scatter(X[30:60, 0], X[30:60, 1], s = 50, c = 'y')
plt.scatter(X[60:90, 0], X[60:90, 1], s = 50, c = 'g')
plt.scatter(X[90:120, 0], X[90:120, 1], s = 50, c = 'c')
plt.scatter(X[120:150, 0], X[120:150, 1], s = 50, c = 'b')

plt.scatter(centers[0,0], centers[0,1], s=200, c='black', marker='s')
plt.scatter(centers[1,0], centers[1,1], s=200, c='black', marker='s')
plt.scatter(centers[2,0], centers[2,1], s=200, c='black', marker='s')
plt.scatter(centers[3,0], centers[3,1], s=200, c='black', marker='s')
plt.scatter(centers[4,0], centers[4,1], s=200, c='black', marker='s')
plt.show()

```

```python
# Testing the algorithm:
# getting the labels property of the K-means clustering example dataset; that is, how the data points are categorized into the two clusters.

Kmean.labels_
# predicting the cluster of a data point
sample_test=np.array([4.0,4.0])

# the test data point belongs to the 0 (green centroid) cluster
second_test=sample_test.reshape(1, -1) # reshape sample_test to a one row (column is set as unspecified value)
Kmean.predict(second_test)
```
