# 劉泰佑 (C100E030)

> Define dataset

```python
X0 = [[-2,-2], [-2,-1], [-1,-1], [0,1], [1,0], [1,1]]
y0 = [1,1,1,2,2,2]
```

> Initialize

```python
import numpy as np

X = np.array(X0)
y = np.array(y0)

print(X)
print(y)
```

> Splitting the data set into the training set and test set

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state = 0)
print ("\nX training set:")
print (X_train)
print ("\nx test set:")
print (X_test)
print ("\ny training set:")
print (y_train)
print ("\ny test set:")
print(y_test)
```

### Fitting SVM to the training set

> Linear Kernel

```python
from sklearn.svm import SVC
clf = SVC (kernel='linear')
clf.fit(X_train, y_train)
```

> Predict the new data set [0,6]

```python
y_pred = clf.predict([[0, 6]])
print (y_pred)
```

> Making the confusion matrix

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print ("\nConfusion matrix:")
print (cm)
```

> Visualizing the training set results - Linear

```python
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt

X_set, y_set = X_train, y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                    np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))

plt.contourf(X1, X2, clf.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
            alpha = 0.3, cmap = ListedColormap(('red', 'green')))

plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())

for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green'))(i), label = j)

plt.title(' SVM (Training set - Linear)')
plt.xlabel ('X0[;,1], i.e Age')
plt.ylabel ('X0[1,:], i.e Estimated Salary')
plt.legend()
plt.show()
```
